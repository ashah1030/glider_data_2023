```python
M5_winds = pd.read_csv('/Users/anjalishah/Desktop/EcoFOCI2024/Data/Mooring_Sites/M5/M5_wnd.2023.csv')
M5_winds.index = pd.to_datetime(M5_winds['time'])
M5_winds = M5_winds.drop(columns = ['time'])

fig, ax = plt.subplots(figsize=(30,5))
x = np.array(M5_winds.index)
u= pd.to_numeric(np.array(M5_winds['uwnd'])) #E/W wind, positive is east
v = pd.to_numeric(np.array(M5_winds['vwnd'])) #N/S wind, positive is north

wind_plot = ax.quiver(x, np.zeros(len(x)), u, v, scale=10, scale_units='inches')
ax.set_title('M5 winds 2023')
ax.quiverkey(wind_plot, 0.07, 0.85, U = 10, label ='10 cm/s', labelpos='E') 
ax.set_yticks([])
#ax.set_ylim(-100, 100)
plt.show()

M5_winds.to_csv('/Users/anjalishah/Desktop/EcoFOCI2024/Data/Mooring_Sites/M5/M5_winds.csv')
# for i, date in (M5_winds.index):
#     x[i] = M5_winds.index[i]
#     y[i] = 0
#     u[i]=  M5_winds['uwnd'][i] #E/W wind, positive is east
#     v[i] = M5_winds['vwnd'][i] #N/S wind, positive is north
#     ax.quiver([x[i], y[i]], u[i], v[i])
```
